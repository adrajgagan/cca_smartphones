```python
import pandas as pd

# Load the dataset
smartphones_df = pd.read_csv('dataset/combined.csv')

# Fill missing values
smartphones_df.fillna('', inplace=True)

# Combine features into a single string
smartphones_df['combined_features'] = smartphones_df.apply(
    lambda row: f"{row['model']} {row['price']} {row['avg_rating']} {row['5G_or_not']} {row['processor_brand']} "
                f"{row['num_cores']} {row['processor_speed']}GHz {row['battery_capacity']}mAh {row['fast_charging_available']} "
                f"{row['fast_charging']} {row['ram_capacity']}GB {row['internal_memory']}GB {row['screen_size']}inch "
                f"{row['refresh_rate']}Hz {row['num_rear_cameras']} {row['os']} {row['primary_camera_rear']}MP "
                f"{row['primary_camera_front']}MP {row['extended_memory_available']} {row['resolution_height']} "
                f"{row['resolution_width']}", axis=1)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Vectorize the combined features
vectorizer = TfidfVectorizer()
feature_matrix = vectorizer.fit_transform(smartphones_df['combined_features'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Function to recommend smartphones
def recommend_smartphone(user_input):
    user_input_vector = vectorizer.transform([user_input])
    similarity_scores = cosine_similarity(user_input_vector, feature_matrix)
    top_indices = similarity_scores[0].argsort()[-5:][::-1]
    return smartphones_df.iloc[top_indices]
```
